```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Datasets/heart-disease.csv")

df.head()
```

```python
over_50 = df[df["age"] > 50]

over_50.head()
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

scat1 = ax0.scatter(over_50["age"], over_50["chol"], c=over_50["target"], cmap="viridis")
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol",
       ylim=(100, 600))
ax0.legend(*scat1.legend_elements(), title="Target", frameon=True, facecolor="white", edgecolor="black")
ax0.axhline(y=over_50["chol"][over_50["target"] == 1].mean(), c="yellow", linestyle='--')
ax0.axhline(y=over_50["chol"][over_50["target"] == 0].mean(), c="purple", linestyle='--')

scat2 = ax1.scatter(over_50["age"], over_50["thalach"], c=over_50["target"], cmap="viridis")
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate",
       ylim=(60, 200))
ax1.legend(*scat2.legend_elements(), title="Target", frameon=True, facecolor="white", edgecolor="black")
ax1.axhline(y=over_50["thalach"][over_50["target"] == 1].mean(), c="yellow", linestyle='--')
ax1.axhline(y=over_50["thalach"][over_50["target"] == 0].mean(), c="purple", linestyle='--')

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
plt.xlim(50, 80)
plt.show()
```
